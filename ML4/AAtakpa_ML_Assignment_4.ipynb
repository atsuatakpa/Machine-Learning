##### Set up the variables we will use.
Create 2000 random x values between -1 and +1 and save them in a vector called X.
Create 2000 random y values between -1 and +1 and save them in a vector called Y.
Use the randomn X and Y values to create a highly non-linear function z=x^5 + y^4 - x^4 - y^3

```python
import numpy as np
x = np.random.uniform(low = -1,high = 1, size=2000)
y = np.random.uniform(low = -1,high = 1, size=2000)
x,y
```

```python
np.shape(x)
```

```python
z=x**5 + y**4 - x**4 - y**3
z
```

```python
Out = z.reshape(2000,1)
Out
```

```python
# Set up the training dataset for the neural network using x & y as the inputs
# saved in a row matrix called In (i.e. the two rows of In are x & y).
# Set up the output row matrix called Out containing z.
# Input
 
# Output (target)
```

```python
In = np.row_stack((x,y))
In= In.reshape(2000,2)
np.shape(In)
```

```python
import pandas as pd
df = pd.DataFrame(In)
df.head(5)
```

```python
# Now do neural network fit using MLPRegressor
# Choose a Training Function
 
# Create a Fitting Network with 10 nodes in the hidden layer
 
# Setup Division of the training Data as follows:
# 85% of the data used for Training,
# 15& of the Data used for Testing
```

```python
from sklearn.neural_network import MLPRegressor
import numpy as np
from sklearn.model_selection import train_test_split
import numpy as np
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt
```

```python
reg = MLPRegressor(hidden_layer_sizes=(10,),  activation='relu', solver='adam',    alpha=0.001,batch_size='auto',
               learning_rate='constant', learning_rate_init=0.01, power_t=0.5, max_iter=1000, shuffle=True,
               random_state=None, tol=0.0001, verbose=False, warm_start=False, momentum=0.9,
               nesterovs_momentum=True, early_stopping=False, validation_fraction=0.1, beta_1=0.9, beta_2=0.999,
               epsilon=1e-08)
```

```python
X_train, X_test, y_train, y_test = train_test_split(Out, In, test_size=0.15, random_state=42)
```

```python
np.shape(y_test)
```

```python
y_test
```

```python
net = reg.fit(X_train,y_train)
```

```python
net
```

```python
pred = net.predict(X_test)
pred
```

```python
# Printing the MSE For Train and Test
print("Mean squared error for Train: %.2f"
      % mean_squared_error(y_test, pred))
```
